```python
# CSV 읽어오기
import pandas as pd

data = pd.read_csv('sample.csv')
```

```python
# 1. 총합 평균 

# 1-1 인덱스 설정
data = data.set_index('학번', drop=True)

# 총합, 평균 구하기 : axis=1이면 각 행의 합, axis=0이면 각 열의 합
sum = data.sum(axis = 1)
mean = data.mean(axis = 1)

# 열 추가
data['총합'] = sum
data['평균'] = mean
print(data)
```

```python
# 2. 총합을 기준으로 순위
data['순위'] = data['총합'].rank()
print(data)
```

```python
# 3. 순위를 기준으로 오름차순
data = data.sort_values(axis=0, by='순위',ascending=True)
print(data)
```

```python
# 4. 등급
def func(row):
  if row > 90:
    return 'A'
  elif row > 80:
    return 'B'
  else:
    return 'C'

data['학점'] = data['평균'].apply(func)
print(data)
```
